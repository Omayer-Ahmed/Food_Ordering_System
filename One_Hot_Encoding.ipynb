```python
import pandas as pd
```

```python
df_csv = pd.read_csv("used_car_cleaned.csv")
```

```python
df_csv.head()
```

```python
df_csv.isnull().sum()
```

```python
new_df = df_csv.drop(["car_model","car_transmission"],axis=1)
```

```python
new_df.head()
```

```python
new_df[new_df["car_brand"] == "Hyundai"]
```

```python
new_df[new_df["car_brand"] == "Chevrolet"]
```

```python
new_df[(new_df["car_brand"] =="Chevrolet") & (new_df["car_price"]>13000)]
```

```python
dummies =pd.get_dummies(new_df["car_brand"])
dummies.head()
```

```python
merge=pd.concat([new_df,dummies],axis='columns')
merge
```

```python
merge[merge["BMW"]==1]
```

```python
original_df=merge.drop(["car_brand","car_model_year"],axis=1)
original_df
```

```python
y=original_df["car_price"]
y
```

```python
x=original_df.drop(["car_price"],axis=1)
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.score(x,y)
```
