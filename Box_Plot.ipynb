#Box plot with matplotlib.pyplot

```python
import matplotlib.pyplot as plt
```

```python
#sample data set
data=[-14,1,2,2,2,3,3,4,5,5,5,6,6,6,6,7,8,8,9,16,27,50]

#create a figure and axis
fig,ax=plt.subplots()

#crate a box plot
ax.boxplot(data)

#add labels and title
plt.xlabel("data")
plt.ylabel("values")
plt.title("Box plot")


#display the plot
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

#sample data
data=[-14,1,2,2,2,3,3,4,5,5,5,6,6,6,6,7,8,8,9,16,27]
sb.boxplot(data=data,color='r')

#define axis and title
plt.xlabel("data")
plt.ylabel("values")
plt.title("Box plot")
plt.show()
```
