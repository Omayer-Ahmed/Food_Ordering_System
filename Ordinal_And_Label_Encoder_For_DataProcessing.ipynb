```python
import pandas as pd
df_csv=pd.read_csv("weather-nominal.csv")
df_csv.head()
```

```python
df_csv["temperature"].unique()
```

```python
order_temp=['hot','mild','cool']
```

# Use of OrdinalEncoder

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
#create object
ore=OrdinalEncoder(categories =[order_temp])
```

```python
ore.fit(df_csv[["temperature"]])
```

```python
new_df = ore.transform(df_csv[["temperature"]])
new_df
```

```python
#replace the new_df into df_csv's temperature
df_csv["temperature"]=pd.DataFrame(new_df)
```

```python
df_csv
```

# Use of LabelEncoder


```python
from sklearn.preprocessing import LabelEncoder
```

```python
le= LabelEncoder()
```

```python
df_csv["outlook"]=le.fit_transform(df_csv["outlook"])
```
