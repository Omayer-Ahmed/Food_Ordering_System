```python
import pandas as pd
df_csv=pd.read_csv("Profit Prediction using Linear Regression.csv")
```

```python
df_csv.head()
```

```python
df_csv.isnull().sum()
```

```python
df_csv["Area"].unique()
```

```python
dummies=pd.get_dummies(df_csv["Area"])
```

```python
dummies.head()
```

```python
merge=pd.concat([df_csv,dummies],axis="columns")
```

```python
merge.head()
```

```python
original_df=merge.drop(["Area"],axis=1)
```

```python
original_df.tail()
```

```python
y=original_df["Profit"]
y.head()
```

```python
x=original_df.drop(["Profit"],axis=1)
```

```python
x.tail()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.predict([[1315.46 ,	115816.21 ,	297114.46 ,	0 ,	0 ,	1]])
```

```python
reg.score(x,y)
```

#Save the model Using Pickle library

```python
import pickle as pk

with open('Model_1','wb') as file:
  pk.dump(reg,file)
```
